Theory Questions & Answers (Statistics Part 2)

Q.1 What is hypothesis testing in statistics?

Answer :-

It's a method to make inferences about population parameters using sample data. It helps you decide whether to accept or reject a claim about a population.

Q.2 What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer :-

Null hypothesis (H₀): Assumes no effect or no difference.
Alternative hypothesis (H₁): Assumes there is an effect or a difference.

We test evidence against the null to support the alternative.

Q.3 What is the significance level in hypothesis testing, and why is it important?

Answer :-

It's the probability of rejecting the null hypothesis when it's actually true (Type I error). Common levels: 0.05, 0.01. It sets the threshold for how strong the evidence must be.

Q.4 What does a P-value represent in hypothesis testing?

Answer :-

It's the probability of getting a test statistic as extreme as the observed one, assuming H₀ is true.


Q.5 How do you interpret the P-value in hypothesis testing?

Answer :-

If P ≤ α: Reject H₀ (evidence supports H₁).
If P > α: Fail to reject H₀ (insufficient evidence for H₁).

Q.6 What are Type 1 and Type 2 errors in hypothesis testing?

Answer :-

Type I Error (α): Rejecting H₀ when it's true.
Type II Error (β): Not rejecting H₀ when it's false.



Q.7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Answer :-

One-tailed: Tests for effect in one direction (e.g., greater than).
Two-tailed: Tests for effect in both directions (e.g., not equal to).

Q.8 What is the Z-test, and when is it used in hypothesis testing?

Answer :-

It's used to compare a sample mean to a known population mean when the population variance is known and sample size is large (n ≥ 30).

Q.9 How do you calculate the Z-score, and what does it represent in hypothesis testing?

Answer :-

Z=Xˉ−μσ/nZ = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
It measures how many standard errors a sample mean is from the population mean.

Q.10 What is the T-distribution, and when should it be used instead of the normal distribution?

Answer :-

Used when sample size is small (n < 30) and population standard deviation is unknown.

Q.11 What is the difference between a Z-test and a T-test?

Answer :-

Z-test: Known population variance, large sample.
T-test: Unknown population variance, small sample.

Q.12 What is the T-test, and how is it used in hypothesis testing?

Answer :-

It tests whether the mean of a sample differs from a known value (one-sample), or whether two means are significantly different (independent or paired).

Q.13 What is the relationship between Z-test and T-test in hypothesis testing?

Answer :-

Both are used for mean comparisons. The T-distribution approaches the Z-distribution as the sample size increases.

Q.14 What is a confidence interval, and how is it used to interpret statistical results?

Answer :-

It gives a range within which the population parameter is likely to lie, with a given level of confidence (e.g., 95%).


Q.15 What is the margin of error, and how does it affect the confidence interval?

Answer :-

It’s the range added/subtracted from the sample mean to create the confidence interval. Larger margin = wider interval.

Q.16 How is Bayes' Theorem used in statistics, and what is its significance?

Answer :-

It updates the probability of a hypothesis given new evidence:
P(A∣B)=P(B∣A)P(A)P(B)P(A|B) = \frac{P(B|A)P(A)}{P(B)}
Useful in machine learning, decision making, and medical testing.

Q.17 What is the Chi-square distribution, and when is it used?

Answer :-

It's used in tests of independence and goodness-of-fit. It deals with categorical data.

Q.18 What is the Chi-square goodness of fit test, and how is it applied?

Answer :-

It checks whether observed categorical data fits an expected distribution.

Q.19 What is the F-distribution, and when is it used in hypothesis testing?

Answer :-

Used to compare variances or test multiple group means (e.g., in ANOVA).

Q.20 What is an ANOVA test, and what are its assumptions?

Answer :-

Tests whether there are differences among group means. Assumes:
o	Normal distribution
o	Homogeneity of variances
o	Independence


Q. 21What are the different types of ANOVA tests?

Answer :-

o	One-way ANOVA
o	Two-way ANOVA
o	Repeated measures ANOVA

Q. 22. What is the F-test, and how does it relate to hypothesis testing?

Answer :-

Compares variances to assess model fit or group differences (as in ANOVA).


```python
# Q. 1. Write a python program to perform a z – test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Known population parameters
population_mean = 100
population_std_dev = 15

# Sample data
sample_data = [102, 98, 101, 97, 105, 100, 99, 103, 96, 104]
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

# Perform z-test
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed test

# Output results
print(f"Sample Mean = {sample_mean}")
print(f"Z-Score = {z_score:.4f}")
print(f"P-Value = {p_value:.4f}")


# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")
```

```python
# Q. 4. Perform a two – tailed z – test using python and visualize the decision region on a plot.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Known population parameters
population_mean = 100
population_std_dev = 15

# Sample data (you can change this or simulate it)
sample_data = [104, 98, 101, 96, 105, 102, 99, 103, 97, 100]
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed test

# Significance level
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

# Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-Score = {z_score:.4f}")
print(f"Z-Critical (±1.96 for alpha=0.05) = ±{z_critical:.4f}")
print(f"P-Value = {p_value:.4f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

# -------------------
# Plotting the decision regions
# -------------------

# X values for standard normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.3, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3, label='Rejection Region (right)')

# Draw Z-score line
plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Mark critical z-values
plt.axvline(-z_critical, color='red', linestyle='dashed', linewidth=1)
plt.axvline(z_critical, color='red', linestyle='dashed', linewidth=1)

# Annotations and labels
plt.title('Two-Tailed Z-Test: Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q. 5. Create a python function that calculates and visualizes type 1 and type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors for hypothesis testing.

    Parameters:
    - mu_0: Mean under null hypothesis
    - mu_1: Mean under alternative hypothesis
    - sigma: Standard deviation of population
    - n: Sample size
    - alpha: Significance level (Type I error rate)
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)
    critical_low = mu_0 - z_critical * se
    critical_high = mu_0 + z_critical * se

    # X range for plotting
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)

    # Distributions under H0 and H1
    y_null = norm.pdf(x, mu_0, se)
    y_alt = norm.pdf(x, mu_1, se)

    # Plot the distributions
    plt.figure(figsize=(12, 6))
    plt.plot(x, y_null, label='H₀ Distribution', color='blue')
    plt.plot(x, y_alt, label='H₁ Distribution', color='green')

    # Shade Type I error region (α): Reject H₀ when H₀ is true
    plt.fill_between(x, y_null, where=(x < critical_low) | (x > critical_high),
                     color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error region (β): Fail to reject H₀ when H₁ is true
    plt.fill_between(x, y_alt, where=(x >= critical_low) & (x <= critical_high),
                     color='orange', alpha=0.3, label='Type II Error (β)')

    # Calculate beta
    beta = norm.cdf(critical_high, mu_1, se) - norm.cdf(critical_low, mu_1, se)

    # Display details
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.axvline(critical_low, color='black', linestyle='--', label='Critical Region')
    plt.axvline(critical_high, color='black', linestyle='--')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical values: [{critical_low:.2f}, {critical_high:.2f}]")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05)
```

```python
# Q. 9 Write a python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - Tuple: (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-value

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Interval ({int(confidence*100)}%): ({lower:.2f}, {upper:.2f})")
    return (lower, upper)

# Example usage:
sample_data = [52, 55, 48, 50, 53, 51, 49, 54, 50, 52]
confidence_interval(sample_data, confidence=0.95)

```

```python
# Q. 10 Write a python program to calculate the margin of error for a given confidence level using sample data.

import math
import scipy.stats as stats

def calculate_margin_of_error(sample_data, confidence_level):
    # Sample statistics
    sample_size = len(sample_data)
    sample_mean = sum(sample_data) / sample_size
    sample_std = math.sqrt(sum((x - sample_mean) ** 2 for x in sample_data) / (sample_size - 1))

    # Critical value (z or t)
    alpha = 1 - confidence_level
    df = sample_size - 1  # degrees of freedom
    t_critical = stats.t.ppf(1 - alpha / 2, df)

    # Margin of Error
    margin_of_error = t_critical * (sample_std / math.sqrt(sample_size))

    return margin_of_error

# Example usage
sample = [45, 50, 55, 47, 53, 49, 52]  # sample data
confidence = 0.95

moe = calculate_margin_of_error(sample, confidence)
print(f"Margin of Error at {confidence*100:.0f}% confidence level: {moe:.2f}")
```

```python
# Q. 11 Implement a Bayesian inference method using bayes’ Theorem in python and explain the process.

def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    - prior_H: P(H), prior probability of the hypothesis
    - likelihood_D_given_H: P(D|H), probability of data given hypothesis is true
    - likelihood_D_given_not_H: P(D|¬H), probability of data given hypothesis is false

    Returns:
    - posterior: P(H|D), updated probability of hypothesis after seeing data
    """
    # Compute P(D)
    prior_not_H = 1 - prior_H
    evidence_D = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)

    # Compute P(H|D)
    posterior = (likelihood_D_given_H * prior_H) / evidence_D
    return posterior

# Given values
P_H = 0.01                # 1% of people have the disease
P_D_given_H = 0.95        # Test correctly identifies disease 95% of the time
P_D_given_not_H = 0.05    # Test falsely identifies disease 5% of the time

posterior = bayes_theorem(P_H, P_D_given_H, P_D_given_not_H)
print(f"Probability of having the disease given a positive test result: {posterior:.4f}")
```

```python
# Q. 12 Perform a chi- square test for independence between two categorical variables in python.

import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
data = [[20, 10],
        [30, 40]]
labels = ['Male', 'Female']
columns = ['Like', 'Dislike']

df = pd.DataFrame(data, index=labels, columns=columns)

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(df)

# Output results
print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=labels, columns=columns))

# Interpret result
alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject the null hypothesis – variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis – variables are independent.")
```

```python
# Q. 13 Write a python program to calculate the expected frequencies for a chi – square text based on observed data.

import pandas as pd

def calculate_expected_frequencies(observed):
    # Convert to DataFrame if input is a list
    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    # Calculate row totals, column totals, and grand total
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()

    # Calculate expected frequencies
    expected = pd.DataFrame(index=observed.index, columns=observed.columns)

    for row in observed.index:
        for col in observed.columns:
            expected.loc[row, col] = (row_totals[row] * col_totals[col]) / grand_total

    return expected.astype(float)

# Example usage:
data = [[20, 10],
        [30, 40]]
observed_df = pd.DataFrame(data, index=['Male', 'Female'], columns=['Like', 'Dislike'])

expected_df = calculate_expected_frequencies(observed_df)
print("Observed Frequencies:\n", observed_df)
print("\nExpected Frequencies:\n", expected_df)
```

```python
# Q. 16. Implement an f – text using python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # F statistic (larger variance in numerator)
    if var1 > var2:
        F = var1 / var2
        dfn, dfd = df1, df2
    else:
        F = var2 / var1
        dfn, dfd = df2, df1

    # Critical values
    critical_value = f.ppf(1 - alpha/2, dfn, dfd)
    lower_critical = f.ppf(alpha/2, dfn, dfd)

    # P-value
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

    print("F Statistic:", F)
    print("Degrees of Freedom:", dfn, "and", dfd)
    print("P-value:", p_value)
    print("Critical Values (Two-tailed):", lower_critical, "to", critical_value)

    # Interpretation
    if p_value < alpha:
        print("\nConclusion: Reject the null hypothesis – variances are significantly different.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis – no significant difference in variances.")

# Example usage:
np.random.seed(42)
sample1 = np.random.normal(10, 2, 30)  # mean=10, std=2
sample2 = np.random.normal(10, 4, 30)  # mean=10, std=4

f_test(sample1, sample2)
```

```python
# Q. 19. Write a python function to check the assumption ( normality, independence and equal variance ) for ANOVA.


import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    # 1. Normality check with Shapiro-Wilk Test
    print("=== Normality Check (Shapiro-Wilk) ===")
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i}: W = {stat:.4f}, p = {p:.4f}")
        if p > alpha:
            print(" -> Normally distributed")
        else:
            print(" -> NOT normally distributed")

    print("\n=== Homogeneity of Variance (Levene’s Test) ===")
    stat, p = levene(*groups)
    print(f"Levene’s test: W = {stat:.4f}, p = {p:.4f}")
    if p > alpha:
        print(" -> Equal variances assumed")
    else:
        print(" -> Variances are significantly different (assumption violated)")

    print("\n=== Independence ===")
    print(" -> Cannot be tested statistically here; assume it's met by design.")
```

```python
group1 = [85, 90, 88, 75, 95]
group2 = [70, 65, 80, 72, 68]
group3 = [90, 92, 88, 94, 89]

check_anova_assumptions(group1, group2, group3)
```

```python
# Q. 20. Perform a two – way ANOVA test using python to study the interaction between two factors and visualize the results.

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Score': [88, 92, 85, 90, 75, 78, 82, 79, 91, 94, 89, 93, 72, 74, 77, 76],
    'Method': ['A']*8 + ['B']*8,
    'StudyTime': ['Short', 'Long']*4 + ['Short', 'Long']*4
}
df = pd.DataFrame(data)

# Two-Way ANOVA
model = ols('Score ~ C(Method) + C(StudyTime) + C(Method):C(StudyTime)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\n=== Two-Way ANOVA Results ===")
print(anova_table)

# ----------- Visualization -----------
plt.figure(figsize=(8, 5))
sns.pointplot(data=df, x='StudyTime', y='Score', hue='Method', markers=['o', 's'], capsize=.1)
plt.title("Interaction Plot: Method vs StudyTime")
plt.ylabel("Average Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q. 21. Write a python program to visualize the F – distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(dfn, dfd, alpha=0.05):
    x = np.linspace(0, 5, 1000)
    y = f.pdf(x, dfn, dfd)

    # Critical F-value
    f_critical = f.ppf(1 - alpha, dfn, dfd)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='steelblue')
    plt.fill_between(x, y, where=(x > f_critical), color='orange', alpha=0.5, label=f'α = {alpha}')

    plt.axvline(f_critical, color='red', linestyle='--', label=f'Critical F = {f_critical:.2f}')
    plt.title('F-Distribution with Shaded Rejection Region')
    plt.xlabel('F value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
plot_f_distribution(dfn=5, dfd=20)
```

```python
# Q. 23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
from scipy.stats import ttest_1samp
import matplotlib.pyplot as plt

# Step 1: Simulate data
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# Step 2: Perform one-sample t-test (null hypothesis: mean = 52)
t_stat, p_value = ttest_1samp(sample_data, popmean=52)

print("=== One-Sample T-Test ===")
print("Sample Mean:", np.mean(sample_data))
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Step 3: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis – the sample mean is significantly different from 52.")
else:
    print("Conclusion: Fail to reject the null hypothesis – no significant difference from 52.")

# Step 4: Visualization
plt.hist(sample_data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(x=52, color='red', linestyle='--', label='Hypothesized Mean (52)')
plt.axvline(x=np.mean(sample_data), color='green', linestyle='--', label=f'Sample Mean ({np.mean(sample_data):.2f})')
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q. 24. Perform a hypothesis test for population variance using a chi – square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

# Step 1: Sample data
np.random.seed(1)
data = np.random.normal(loc=50, scale=8, size=30)  # Normal with true std dev = 8
sample_var = np.var(data, ddof=1)
n = len(data)

# Step 2: Hypothesized population variance
sigma_0_squared = 64  # Hypothesized variance (8^2)

# Step 3: Test statistic
chi_square_stat = (n - 1) * sample_var / sigma_0_squared

# Step 4: Critical values and p-value
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(
    chi2.cdf(chi_square_stat, df),
    1 - chi2.cdf(chi_square_stat, df)
)

# Step 5: Print results
print("=== Chi-Square Test for Variance ===")
print(f"Sample Variance: {sample_var:.2f}")
print(f"Chi-Square Statistic: {chi_square_stat:.2f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Values (α={alpha}): [{chi2_lower:.2f}, {chi2_upper:.2f}]")
print(f"P-value: {p_value:.4f}")

# Step 6: Conclusion
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Conclusion: Reject the null hypothesis – variance is significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis – no significant difference in variance.")
```

```python
# Q. 25. Write a python script to perform a z – test for comparing proportions between two datasets or groups.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Success counts and sample sizes for two groups
success_counts = np.array([50, 30])     # Number of "successes" (e.g., clicks)
n_obs = np.array([200, 180])            # Total observations per group

# Perform the Z-test
z_stat, p_value = proportions_ztest(count=success_counts, nobs=n_obs)

# Print results
print("=== Z-Test for Two Proportions ===")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis – proportions are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis – no significant difference in proportions.")
```

```python
# Q. 26. Implement an f- test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Generate two random datasets
np.random.seed(0)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=52, scale=15, size=35)

# Step 2: Calculate variances and F-statistic
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# Ensure the larger variance is in numerator
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(group1) - 1, len(group2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(group2) - 1, len(group1) - 1

# Step 3: Compute p-value for two-tailed test
p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

# Step 4: Print results
print("=== F-Test for Equality of Variances ===")
print(f"Variance Group 1: {var1:.2f}")
print(f"Variance Group 2: {var2:.2f}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"Degrees of Freedom: ({dfn}, {dfd})")
print(f"P-Value: {p_value:.4f}")

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis – variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis – no significant difference in variances.")

# Step 6: Visualization of F-distribution
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-stat = {f_stat:.2f}')
plt.title('F-Test: Comparing Variances')
plt.xlabel('F Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q. 27. Perform a chi – square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate die rolls
np.random.seed(1)
rolls = np.random.randint(1, 7, size=120)  # 120 rolls of a fair die

# Step 2: Count observed frequencies
observed_freq = np.bincount(rolls)[1:]  # index 0 is unused (since die faces start from 1)
expected_freq = np.array([20] * 6)      # expected = 120 rolls / 6 faces = 20 per face

# Step 3: Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_freq, f_exp=expected_freq)

# Step 4: Print results
print("=== Chi-Square Goodness-of-Fit Test ===")
print("Observed Frequencies:", observed_freq)
print("Expected Frequencies:", expected_freq)
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis – the die may not be fair.")
else:
    print("Conclusion: Fail to reject the null hypothesis – the die appears fair.")

# Step 6: Visualization
categories = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(categories))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, observed_freq, width=0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected_freq, width=0.4, label='Expected', color='orange')
plt.xticks(x, categories)
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies (Chi-Square Test)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
